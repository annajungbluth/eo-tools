### Randomly sample 50000 files from the MSG archive

```python
def random_date(start, end):
    """
    Generate a random datetime between two datetime objects.
    """
    delta = end - start
    random_days = np.random.randint(0, delta.days + 1)
    return start + timedelta(days=random_days)

def random_time(start, end):
    """
    Generate a random time between two time objects.
    """
    start_minutes = (start.hour * 60) + (start.minute)
    end_minutes = (end.hour * 60) + (end.minute)
    random_minutes = np.random.randint(start_minutes, end_minutes + 1)
    return datetime(2000, 1, 1, random_minutes // 60, random_minutes % 60, 0)

def random_datetime(start, end):
    """
    Generate a random datetime between two datetime objects.
    """
    random_date_value = random_date(start, end)
    random_time_value = random_time(datetime(2000, 1, 1, 0, 0, 0), datetime(2000, 1, 1, 23, 59, 00))
    return datetime(random_date_value.year, random_date_value.month, random_date_value.day,
                    random_time_value.hour, random_time_value.minute, random_time_value.second)
```

```python
def sample_msg_files(start, end, num_files=10):
    """
    Sample random MSG files from within the specified date range.
    Compile a dataframe with the samples filenames.
    Args:
        start (datetime): Start date for sampling.
        end (datetime): End date for sampling.
        num_files (int): Number of files to sample.
    """

    sampled_files = []

    # add progress bar
    pbar = tqdm(total=num_files, desc="Sampling MSG files", unit="file")
    pbar.set_postfix(start=start, end=end)
    pbar.update(0)
    while len(sampled_files) < num_files:
        try:
            dt = random_datetime(start, end)
            year = dt.year
            month = dt.month
            day = dt.day
            hour = dt.hour

            # list all files available for this date and hour
            seviri_files = os.listdir(f'{FILEPATH}/{year}/{month:02d}/{day:02d}/')

            hour_files = [f for f in seviri_files if f'NA-{year}{month:02d}{day:02d}{hour:02d}' in f]
            
            if not hour_files:
                continue  # Skip if no files found for this datetime
            
            sampled_file = np.random.choice(hour_files)
            sampled_files.append({
                'filename': str(sampled_file),
                'path': f'{FILEPATH}/{year}/{month:02d}/{day:02d}/{sampled_file}',
                'datetime': dt,
            })
            pbar.update(1)
            
        except Exception as e:
            print(f"Error sampling file for datetime {dt}: {e}")
            continue
    
    # turn the list of sampled files into a DataFrame
    df = pd.DataFrame(sampled_files)
    df['datetime'] = pd.to_datetime(df['datetime'])
    pbar.close()
    return df
```

```python
import sys
import pathlib
sys.path.append(str(pathlib.Path().resolve().parent))

from scripts.process_utils import CenterWeightedCropDatasetEditor, read_zipped_msg

def load_and_patch_msg(file_path, patch_size, fov_radius):
    """
    Load and patch MSG file using Satpy.
    
    Args:
        file_path (str): Path to the MSG file.
        patch_size (int): Size of the patch to crop from the dataset.
        fov_radius (float): Field of view radius for cropping.
    
    Returns:
        xarray.Dataset: Patched dataset.
    """
    ds = read_zipped_msg(file_path)

    # Crop dataset into patch
    crop = CenterWeightedCropDatasetEditor(
        patch_shape=(patch_size, patch_size), 
        fov_radius=fov_radius,
        satellite = 'msg')
    
    result = crop(ds)

    if result is None: # i.e. if no valid patch was found
        print(f"Could not find valid patch ...")
        return None

    return result
```

```python
def check_quality_flags_msg(ds, cutoff=10000):
    """
    Function to check quality in MSG data.

    Args:
        ds (xarray.Dataset): The dataset to check.
        cutoff (int): The maximum number of NaN values allowed. If exceeded, the function will return False. 
            There will be NaN values close to the disk edge, even after filtering for valid lat/lon values.
            From experimentation, these cases might have around 8000 to 10000 NaN values close to the disk edge
            even if the data on disk is valid. To not filter out all edge disk images, we emperically set the default cutoff to 10000.
    Returns:
        bool: True if the dataset passes the quality check, False otherwise.
    """
    channels=[
            'IR_016',
            'IR_039',
            'IR_087',
            'IR_097',
            'IR_108',
            'IR_120',
            'IR_134',
            'VIS006',
            'VIS008',
            'WV_062',
            'WV_073']
    # create a mask where the latitude value is inf
    mask_lat = ~np.isinf(ds.latitude)
    mask_lon = ~np.isinf(ds.longitude)
    # combine both masks to find all points with valid lat/lon values
    mask = mask_lat & mask_lon
    # loop through each channel to check for NaN values
    for channel in channels:
        # Check for NaN only where mask is True, i.e. where the lat/lon values are valid
        nan_in_valid_region = np.isnan(ds[channel].values[mask])
        if np.count_nonzero(nan_in_valid_region) > cutoff:
            return False
        else: 
            return True
```

```python
def get_cloud_file(date, base_path="/gws/ssde/j25b/eo_shared_data/vol1/satellite/seviri-orac/Data"):
    """
    Get the cloud file for a given date.
    
    Args:
        date (datetime): The date for which to get the cloud file.
        base_path (str): The base path where the files are stored.
    
    Returns:
        str: The path to the cloud file.
    """
    date_str = date.strftime("%Y%m%d%H%M%S")
    cloud_files = os.listdir(f"{base_path}/{date.year}/{date.month:02d}/{date.day:02d}/{date.hour:02d}/")
    relevant_file = [os.path.join(f"{base_path}/{date.year}/{date.month:02d}/{date.day:02d}/{date.hour:02d}/", f) for f in cloud_files if f"{date_str}-ESACCI-L2_CLOUD-CLD" in f and not 'SECONDARY' in f]
    
    if len(relevant_file) == 0:
        print(f"No cloud file found for {date_str}")
        return None
    elif len(relevant_file) > 1:
        raise ValueError(f"Multiple cloud files found for {date_str}: {relevant_file}")
    else:
        return relevant_file[0]
```

```python
def get_seviri_file(date, base_path="/gws/ssde/j25b/eo_shared_data/vol2/satellite/seviri/FDS/"):
    """
    Get the SEVIRI file for a given date.
    
    Args:
        date (datetime): The date for which to get the SEVIRI file.
        base_path (str): The base path where the files are stored.
    
    Returns:
        str: The path to the SEVIRI file.
    """
    date_str = date.strftime("%Y%m%d%H%M%S")
    seviri_files = os.listdir(f"{base_path}/{date.year}/{date.month:02d}/{date.day:02d}/")
    seviri_files = [os.path.join(f"{base_path}/{date.year}/{date.month:02d}/{date.day:02d}/", f) for f in seviri_files]

    seviri_dates = [f.split("/")[-1].split(".")[0].split("-")[-1] for f in seviri_files]
    seviri_dates = [datetime.strptime(dt, "%Y%m%d%H%M%S") for dt in seviri_dates]

    # find closest match
    closest_date = min(seviri_dates, key=lambda x: abs(x - date))

    timedelta_ = abs(closest_date - date)
    if timedelta_ > timedelta(minutes=5):
        # print(f"No SEVIRI file found within 5 minutes of {date_str}. Closest file is {closest_date} with a time difference of {timedelta_}.")
        return None
    else:
        closest_file = seviri_files[seviri_dates.index(closest_date)]
        return closest_file
```

```python
list_dates = []
list_cld = []
list_seviri = []
all_available_files = []

l2_base_path = "/gws/ssde/j25b/eo_shared_data/vol1/satellite/seviri-orac/Data/"
l1_base_path = "/gws/ssde/j25b/eo_shared_data/vol2/satellite/seviri/FDS"

for i in tqdm(range(len(dates['dates']))):
    # print(f"{i}: {dates['dates'][i]}")

    list_dates.append(dates['dates'][i])

    cloud_file = get_cloud_file(dates['dates'][i], base_path=l2_base_path)
    list_cld.append(cloud_file)

    seviri_file = get_seviri_file(dates['dates'][i], base_path=l1_base_path)
    list_seviri.append(seviri_file)

    if cloud_file is not None and seviri_file is not None:
        all_available_files.append(True)
    else:
        all_available_files.append(False)

df = pd.DataFrame({
    'date': list_dates,
    'cloud_file': list_cld,
    'seviri_file': list_seviri,
    'all_available': all_available_files
})
```
